```python
#kiosk v11.7
class Kiosk: 
    def __init__(self):
        # 다차원 리스트로 변경, [메뉴이름, 가격] 묶어서 관리
        # menu[i][0] = 메뉴 이름, menu[i][1] = 가격
        self.menu = [
            ['아메리카노', 2000],
            ['카페라떼  ', 3000],
            ['모카 커피 ', 3000],
            ['유자차    ', 2500],
            ['녹차      ', 2500],
            ['코코아    ', 3000],
        ]

    def menu_print(self):
        print('=' * 30)
        print('      oieh Caffe Menu')
        print('=' * 30)
        print('  [Coffee]')
        for i, item in enumerate(self.menu[:3], start=1):
            # enumerate() 로 인덱스랑 값 동시에 뽑음
            # start=1 로 번호 1부터 시작, 안 하면 0부터 시작해서 찐빠남
            # item[0] = 이름, item[1] = 가격
            print(f'  {i}. {item[0]}  {item[1]}원')
        print('-' * 30)
        print('  [Tea & else]')
        for i, item in enumerate(self.menu[3:], start=4):
            # self.menu[3:] 으로 논커피만 슬라이싱, start=4 로 번호 4부터 시작
            print(f'  {i}. {item[0]}  {item[1]}원')
        print('=' * 30)

    def menu_select(self):
        print()
        self.order_menu = []   # 주문 리스트, 빈 리스트로 시작
        self.order_price = []  # 가격 리스트, 빈 리스트로 시작

        n = 0
        # n 이 1보다 작거나 메뉴 개수 초과하면 계속 다시 받음
        while n < 1 or len(self.menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))

            if 1 <= n and n <= len(self.menu):
                self.order_price.append(self.menu[n-1][1])  # menu[n-1][1] 이 가격임
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 첫 번째 음료 온도 선택
        t = 0  # 기본값 0으로 시작
        while t != 1 and t != 2:
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp = "ICE"
            else:
                print("1과 2 중 하나를 입력하세요.\n")
                t = 0       # 이상한 값 남아있으면 찐빠나니까 초기화
                continue    # while 처음으로 돌아가서 다시 입력받음

        # 온도 + 메뉴 이름 주문 리스트에 추가
        self.order_menu.append(self.temp + ' ' + self.menu[n-1][0])  # menu[n-1][0] 이 이름임

        # 추가 주문 받기, 0 입력하면 주문 완료
        n = 1  # while 바로 안 끝나게 1로 시작
        while n != 0:
            n = int(input("추가 주문은 번호를, 주문 완료는 0을 입력하세요 : "))

            if n == 0:
                break  # 0 입력하면 추가 주문 종료

            elif 1 <= n and n <= len(self.menu):
                self.order_price.append(self.menu[n-1][1])  # 추가 음료 가격도 리스트에 추가

                # 추가 주문도 온도 물어봐야 함, 위랑 똑같은 구조
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                        t = 0       # 초기화
                        continue    # while 처음으로

                self.order_menu.append(self.temp + ' ' + self.menu[n-1][0])  # 추가 음료도 주문 리스트에 추가

            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    def table(self):
        print()
        print('=' * 35)
        print('         oieh Caffe Menu')
        print('         [ 주  문  표 ]')
        print('=' * 35)

        # order_menu 랑 order_price 인덱스 맞춰서 출력
        # range(len()) 활용
        for i in range(len(self.order_menu)):
            print(f'  {self.order_menu[i]:<20} {self.order_price[i]}원')
           
        print('-' * 35)  # 세로줄(|) 쓰면 한글 때문에 오와열 찐빠나서 그냥 뺌
        print(f'  합계 금액 : {sum(self.order_price)}원')  # sum() 으로 전체 가격 합산
        print('=' * 35)
        print('       이용해 주셔서 감사합니다.')
        print('=' * 35)

    def pay(self):
        p = 0  # 기본값 0으로 시작, 1이나 2 입력할 때까지 while 돔
        while p != 1 and p != 2:
            p = int(input("결제 수단을 선택하세요. 현금은 1을, 카드는 2를 입력하세요 : "))

            if p == 1:
                print("직원을 호출하겠습니다.")  # 현금이면 직원 호출
            elif p == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")  # 카드면 꽂으라고 안내
            else:
                print("다시 결제를 시도해 주세요.")
                p = 0       # 이상한 값 남아있으면 찐빠나니까 초기화
                continue    # while 처음으로 돌아가서 다시 입력받음


def run():  # 매번 테스트할때마다 쳐야되니까 런함수에 묶음
    a = Kiosk()
    a.menu_print()
    a.menu_select()
    a.table()
    a.pay()
```
