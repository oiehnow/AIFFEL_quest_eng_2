```python
class Square:
    def __init__(self):
        # input() 은 무조건 str 로 들어오니까 int() 로 감싸줘야 숫자 비교 가능
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        # self.square 에 저장된 숫자에 따라 분기
        # 1이면 rect(), 2면 par(), 3이면 trape() 쓰면 됨
        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')
        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        else:
            # 1,2,3 외에 다른 거 입력하면 여기로 빠짐
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        # 가로, 세로 따로따로 input 으로 받음
        width = int(input('가로를 입력하세요 >>> '))    # str 로 들어오니까 int() 변환
        vertical = int(input('세로를 입력하세요 >>> ')) # 세로도 마찬가지
        area = width * vertical   # 직사각형 넓이 공식 : 가로 x 세로
        result = '직사각형의 넓이 : ' + str(area)  # area 는 int 라서 str() 로 변환해야 문자열이랑 합쳐짐
        return result  # print 아니고 return, 나중에 print(a.rect()) 로 출력

    def par(self):
        # 평행사변형은 밑변이랑 높이 두 개만 있으면 됨
        base = int(input('밑변을 입력하세요 >>> '))    # 밑변
        height = int(input('높이를 입력하세요 >>> '))  # 높이
        area = base * height   # 평행사변형 넓이 공식 : 밑변 x 높이
        result = '평행사변형의 넓이 : ' + str(area)
        return result

    def trape(self):
        # 사다리꼴은 밑변, 윗변, 높이 세 개 다 필요함
        base = int(input('밑변을 입력하세요 >>> '))    # 밑변
        top = int(input('윗변을 입력하세요 >>> '))     # 윗변
        height = int(input('높이를 입력하세요 >>> '))  # 높이
        area = (base + top) * height / 2  # 사다리꼴 넓이 공식 : (밑변 + 윗변) x 높이 / 2
        result = '사다리꼴의 넓이 : ' + str(area)
        return result

a = Square() # 객체 생성, 이 순간 __init__ 실행되면서 숫자 입력받음
```

```python
a.par()
```

```python
a.trape()
```
