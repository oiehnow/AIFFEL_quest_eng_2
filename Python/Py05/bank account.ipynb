Q1. Account 클래스 : 은행에 가서 계좌를 개설하면 은행이름, 예금주, 계좌번호, 잔액이 설정됩니다. Account 클래스를 생성한 후 생성자(hint: 매직메서드..!!)를 구현해보세요.
    생성자에서는 예금주와 초기 잔액만 입력 받습니다. 은행이름은 SC은행으로 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성됩니다. (은행이름: SC은행, 계좌번호: 111-11-111111)

```python
# bank 1.4
import random  # 랜덤 숫자는 무조건 이거

class Account:
    
    def __init__(self, owner, balance):  # 생성자 복습  __init__ 이거
        self.bank_name = 'TERA은행' # 은행이름은 무조건 TERA은행으로 할거니까 이렇게 고정
        
        self.owner = owner # 예금주는 밖에서 넣어주는 값이니까 받아서 세이브
        
        self.balance = balance # 잔액도 밖에서 넣어주는 값 그대로 세입
        
        # 계좌번호 랜덤 생성 3-2-6 
        # random.randint(100, 999) 이러면 100~999 사이 숫자 나옴 즉 3자리 개날먹
        num1 = random.randint(100, 999)        # 3자리
        num2 = random.randint(10, 99)          # 2자리
        num3 = random.randint(100000, 999999)  # 6자리
        
        self.account_num = f'{num1}-{num2}-{num3}' # f스트링 사용, 위에서 만든 숫자들을 - 로 연결해서 계좌번호 만들기
```

```python
# 테스트
a = Account('아라무스', 10000)

print(a.bank_name)    # 은행이름
print(a.owner)        # 고갱 이름
print(a.balance)      # 돈
print(a.account_num)  # 계좌번호
```

```python
# bank 3.2
import random  # 랜덤 숫자는 무조건 이거

class Account:
    account_count = 0  # q2 추가, 클라스 변수, 계좌 세는 용도
    
    def __init__(self, owner, balance):  # 생성자 복습  __init__ 이거
        self.bank_name = 'TERA은행'  # 은행이름은 무조건 TERA은행으로 할거니까 이렇게 고정
        
        self.owner = owner  # 예금주는 밖에서 넣어주는 값이니까 받아서 세이브
        
        self.balance = balance  # 잔액도 밖에서 넣어주는 값 그대로 세입
        
        # 계좌번호 랜덤 생성 3-2-6
        # random.randint(100, 999) 이러면 100~999 사이 숫자 나옴 즉 3자리 개날먹
        num1 = random.randint(100, 999)        # 3자리
        num2 = random.randint(10, 99)          # 2자리
        num3 = random.randint(100000, 999999)  # 6자리
        
        self.account_num = f'{num1}-{num2}-{num3}'  # f스트링 사용, 위에서 만든 숫자들을 - 로 연결해서 계좌번호 만들기
        
        # 계좌가 새로 만들어질 때마다 account_count를 1씩 늘려줘야함
        Account.account_count += 1  # q2 추가, 클래스 변수는 클래스이름.변수명 으로 접근한다고 함 헷갈림 주의, 객체 하나 만들어질때마다 +1

    # q3 추가
    def get_account_num(self):  # 계좌 개수 출력하는 메서드래 이름이 좀 헷갈림 account_num이랑 다른거임 주의
        print(f'현재까지 생성된 계좌 수 : {Account.account_count}개')  # 개수 출력

    # q4 추가
    def deposit(self, amount):  # 입금 메서드 amount는 입금할 금액임 밖에서 넣어주는 값
                                # 최소 1원 이상만 가능하다고 했으니까 조건 체크 해야함
        if amount < 1:  # 트롤링 방지, 입금액이 1원보다 작으면 안된다고 했으니까
                        # 0원이나 마이너스 금액 입금하려는거 막아야함
            print('입금은 최소 1원 이상만 가능하다.')  # 트롤링 방지 문구
        
        else:  # 1원 이상이면 정상적으로 입금 처리
            self.balance += amount  # 잔액에 입금액 더해주기 이게 핵심임
                                    # self.balance = self.balance + amount 랑 똑같은말임
            print(f'{amount}원 입금 성공. 현재 잔액 : {self.balance}원')

    # q5 추가
    def withdraw(self, amount):  # deposit이랑 반대 개념
                                 # amount는 출금할 금액 밖에서 넣어주는 값
                                 # 잔고 이상으로는 출금 못한다고 했으니까 조건 체크 필요
        
        if amount > self.balance:  # 트롤링 방지, 출금하려는 금액이 잔액보다 크면 안됨
                                   # 잔액이 1000원인데 5000원 빼려고 하면 막아야함
                                   # 현실에서도 당연히 안되는거 그걸 코드로 구현하는거임
            print('잔액이 부족합니다.')  # 트롤링 경고 문구
            print(f'현재 잔액 : {self.balance}원')  # 현재 얼마 있는지도 알려주면 친절할것같아서 추가함
        
        else:  # 잔액이 충분하면 정상적으로 출금 처리
            self.balance -= amount  # 잔액에서 출금액 빼주기
                                    # self.balance = self.balance - amount 랑 똑같은말임
                                    # deposit이랑 반대로 += 아니고 -= 임 헷갈리지말기
            print(f'{amount}원 출금 성공. 현재 잔액 : {self.balance}원')
```

```python
a = Account('아라무스', 15000)
a.withdraw(3000) # 3천 출금 기대값 12000원
```

```python
a.withdraw(99999) # 트롤링
```

```python
# bank 3.3
import random  # 랜덤 숫자는 무조건 이거

class Account:
    account_count = 0  # q2 추가, 클라스 변수, 계좌 세는 용도
    
    def __init__(self, owner, balance):  # 생성자 복습  __init__ 이거
        self.bank_name = 'TERA은행'  # 은행이름은 무조건 TERA은행으로 할거니까 이렇게 고정
        
        self.owner = owner  # 예금주는 밖에서 넣어주는 값이니까 받아서 세이브
        
        self.balance = balance  # 잔액도 밖에서 넣어주는 값 그대로 세입
        
        # 계좌번호 랜덤 생성 3-2-6
        # random.randint(100, 999) 이러면 100~999 사이 숫자 나옴 즉 3자리 개날먹
        num1 = random.randint(100, 999)        # 3자리
        num2 = random.randint(10, 99)          # 2자리
        num3 = random.randint(100000, 999999)  # 6자리
        
        self.account_num = f'{num1}-{num2}-{num3}'  # f스트링 사용, 위에서 만든 숫자들을 - 로 연결해서 계좌번호 만들기
        
        # 계좌가 새로 만들어질 때마다 account_count를 1씩 늘려줘야함
        Account.account_count += 1  # q2 추가, 클래스 변수는 클래스이름.변수명 으로 접근한다고 함 헷갈림 주의, 객체 하나 만들어질때마다 +1

    # q3 추가
    def get_account_num(self):  # 계좌 개수 출력하는 메서드래 이름이 좀 헷갈림 account_num이랑 다른거임 주의
        print(f'현재까지 생성된 계좌 수 : {Account.account_count}개')  # 개수 출력

    # q4 추가
    def deposit(self, amount):  # 입금 메서드 amount는 입금할 금액임 밖에서 넣어주는 값
                                # 최소 1원 이상만 가능하다고 했으니까 조건 체크 해야함
        if amount < 1:  # 트롤링 방지, 입금액이 1원보다 작으면 안됨
            print('입금은 최소 1원 이상만 가능하다.')  # 트롤링 방지 문구
        
        else:  # 1원 이상이면 정상적으로 입금 처리
            self.balance += amount  # 잔액에 입금액 더해주기 이게 핵심임
                                    # self.balance = self.balance + amount 랑 똑같은말임
            print(f'{amount}원 입금 성공. 현재 잔액 : {self.balance}원')

    # q5 추가
    def withdraw(self, amount):  # deposit이랑 반대 개념
                                 # amount는 출금할 금액 밖에서 넣어주는 값
                                 # 잔고 이상으로는 출금 못한다고 했으니까 조건 체크 필요
        if amount > self.balance:  # 트롤링 방지, 출금하려는 금액이 잔액보다 크면 안됨
            print('잔액이 부족합니다.')  # 트롤링 방지 문구
            print(f'현재 잔액 : {self.balance}원')  # 현재 얼마 있는지도 알려주면 친절할것같아서 추가함
        
        else:  # 잔액이 충분하면 정상적으로 출금 처리
            self.balance -= amount  # 잔액에서 출금액 빼주기
                                    # self.balance = self.balance - amount 랑 똑같은말임
                                    # deposit이랑 반대로 += 아니고 -= 임 헷갈리지말기
            print(f'{amount}원 출금 성공. 현재 잔액 : {self.balance}원')

    # q6 추가
    def display_info(self):  # 계좌 정보 출력 메서드
                             # self.뭐시기 들 죄다 출력하면 됨
                             # 근데 잔고는 세자리마다 쉼표 찍어야한다고 했음 어떻게하지
                             # 찾아보니까 f스트링에서 {:,} 이렇게 쓰면 된다고 함
                             # 예를들어 f'{10000:,}' 이러면 10,000 이렇게 나옴 섹스!
        print(f'은행이름: {self.bank_name}')
        print(f'예금주: {self.owner}')
        print(f'계좌번호: {self.account_num}')
        print(f'잔고: {self.balance:,}원')  # {:,} 이거 숨은문제임 숫자에 쉼표 자동으로 찍어줌, 10000 → 10,000
```

```python
# 테스트
a = Account('아라무스', 10000)
a.display_info()
```
